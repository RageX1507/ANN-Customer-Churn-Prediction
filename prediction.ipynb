```python
import tensorflow as tf 
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
##load the model and pickle file
model = load_model('model.h5')

with open('label_encoder_gender.pkl','rb') as file:
    label_gender = pickle.load(file)
    
with open('onehot_encoder_geo.pkl','rb') as file:
    ohe = pickle.load(file)
    
with open('scaler.pkl','rb') as file:
    scaler = pickle.load(file)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
type(input_data)
```

```python
geo_encoded = ohe.transform([[input_data['Geography']]])
geo_encoder_df = pd.DataFrame(geo_encoded,columns=ohe.get_feature_names_out(['Geography']))
geo_encoder_df
```

```python
input_df=pd.DataFrame([input_data])
input_df = pd.concat([input_df.drop('Geography', axis=1), geo_encoder_df], axis=1)
```

```python
input_df['Gender'] = label_gender.transform(input_df['Gender'])
```

```python
input_df
```

```python
##scaile the input data
input_scaled = scaler.transform(input_df)
```

```python
model.predict(input_scaled)
```
